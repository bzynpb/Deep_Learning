#### Cost Function vs Loss Function 

**Cost func.** ve **Loss fonc.** hemen hemen ayni anlama gelir. 
- Loss func.  single training icin gecerlidir, her durumda hesaplanir. (Loss function is usually a function defined on a data point, prediction and label, and measures the penalty.)
- Cost Func. average of the loss functions of the entire training set. (Cost function is usually more general. It might be a sum of loss functions over your training set plus some model complexity penalty (regularization) )
( A loss function/error function is for a single training example/input. A cost function, on the other hand, is the average loss over the entire training dataset.)

#### Epoch & Batchsize Hyperparameters

-  **Batchsize**, gradient descent hyperparametresidir.
- Modele ayni anda kac adet observation egitime girecek, onu belirliyor. Tum networku/egitimi bastan sonra tamamliyor.

- Model icindeki parametreleri guncellemeden once, training edecegi orneklerin sayisini kontrol eder. 
`The batch size is a hyperparameter of gradient descent that controls the number of training samples to work through before the model’s internal parameters are updated.`
- tek islemde, training edecegi observation pointlerin sayisi 
- **Epoch**, gradient descent hyperparametresidir. Training datasetinde, gecislerini tamamlayan datasetlerin sayisini kontrol eder.
`The number of epochs is a hyperparameter of gradient descent that controls the number of complete passes through the training dataset.`

## Backpropagation

uygun agirlik ve biaslari bularak modelin cost func.ini minimize ettik. Bu sayede modeldeki hatalari minimize ederek ogrenim saglaniyor. 
- Gradient descent de hatalari minimize etmek icin kullandigimiz bir optimizasyon algoritmasi. 
- Agirlik ve biaslari guncellemek icin geriye dogru hareket etmemiz gerekiyor. 
- Geriye dogru yapilan tum bu hareketler ve bunun sonucunda agirlik ve biaslarin guncellenmesi **Backpropagation** olarak adlandiriliyor. 

----
- Q: What is "Backpropagation" ?
- A: Backpropagation, multilayer neurol networklerde kullanilan training algorithm. Networkun sonunda bulunan "error bilgisini", network icindeki tum agirliklara gondererek gradient hesaplanmasini saglar. 
    - Output olusturmek icin, network uzerinden training data yayilmasini saglar. `It can forward propagation of training data through the network to generate output.` 
    - Outputta hata turevini hesaplamak icin Target ve output degerlerini kullanir. `It uses target value and output value to compute error derivative concerning output activations.` 
    - Bir onceki layer ve hidden layerlarda, output hareketlerinde hata turevlerini hesaplayarak geriye dogru ilerler.` It can backpropagate to compute the derivative of the error concerning output activations in the previous layer and continue for all hidden layers.` 
    - Ilgili agirliga ait hata turevini hesaplamak icin, output ve tum hidden layerlarda onceden hesaplanan turevleri kullanir ` It uses the previously calculated derivatives for output and all hidden layers to calculate the error derivative concerning weights.`
    -  Agirliklari gunceller`It updates the weights.`
